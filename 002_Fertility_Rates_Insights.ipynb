```python
#Step 1. Import python data libraries to conduct data analysis and visualizations.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
```

```python
# Step 2. Label and Import .csv file
file_path = '/content/Fertility rates_Test2.csv'
df = pd.read_csv(file_path)
```

```python

#Step 3. Inspect the size of the dataset (total number of rows and columns) which is an essential part of the initial exploratory data analysis.

print(df.shape)
```

```python
# Step 4. Check the first 5 rows of the dataset for a quick overview, data content validation, identify anomalies, and column inspection.

df.head(5)
```

```python
# Step 5. Apply Transpose operation on the dataframe

df = df.drop(columns = ["Country Code"])

# Apply Transpose
df = df.T

# Change the column names
df.columns = df.iloc[0]

# Remove the first row as it contains columns names
df = df.iloc[1:]

# Change the index values
df.index = pd.RangeIndex(start=1960, stop=2023, name="Year")
```

```python
# Step 6. Check the first 5 rows of the dataset for a quick overview, data content validation, identify anomalies, and column inspection.

df.head(5)
```

```python
# Step 5. Check the last 5 rows of the dataset to ensure data completenes, and to detect patterns or issues. Both head and tail help in checking the quality of the dataset.

df.tail(5)
```

```python
#Step 6. Get a concise summary of the database that includes the number of non-null entries, data type of columns and memory usage.

print(df.info())
```

```python
# Step 7. Check for the missing data in the dataframe.It helps to identify where data cleaning or imputation is needed.

temp_df = df.isnull().sum(axis=0).reset_index()
temp_df.columns = ["Column Name", "Number of missing data points"]
temp_df
```

```python
# Step 8. Calculate the percentage of missing values in each column.

temp_df = df.isnull().sum(axis=0).reset_index()
temp_df.columns = ["Column Name", "Number of missing data points"]
temp_df["Percentage of missing data points"] = temp_df["Number of missing data points"].apply(lambda x: np.round(x * 100 / df.shape[0], 2))
temp_df.drop(columns=["Number of missing data points"], inplace=True)
print(temp_df)
```

```python
# Step 9. Check for any outliers or extreme values in the dataset that might be a data entry error.

# important_features = ["Country Name", "Year", "United States", "Canada" ]

plt.figure(figsize=(6, 4))
sns.histplot(df['Canada'], kde=True, bins=10, color='teal', alpha=0.7)

# Titles and labels
plt.title('Distribution of Fertility Rates in Canada', fontsize=16)
plt.xlabel('Fertility Rate', fontsize=12)  # X-axis should represent fertility rates
plt.ylabel('Frequency', fontsize=12)  # Y-axis represents count of occurrences

# Show plot
plt.show()
```

```python
# Select data for Canada and US
df_selected = df[['Canada', 'United States']]

# Plot fertility rates across years
plt.figure(figsize=(9, 5))
plt.plot(df_selected.index, df_selected['Canada'], label='Canada', marker='o', linestyle='-')
plt.plot(df_selected.index, df_selected['United States'], label='United States', marker='s', linestyle='--')

# Formatting
plt.title('Fertility Rate Comparison: Canada vs. US (1960-2022)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Fertility Rate', fontsize=12)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# Fertility Rate Decline After Policy Changes
# Assessing the impact of policies like China’s One-Child Policy or government incentives for childbirth such as in France.
# Compare fertility rates before & after major policy changes. France offers various government incentives for childbirth, including a birth grant (around €950), monthly child benefits, and diverse family allowances, along with maternity leave and other support measures.
# China's One-Child Policy was officially implemented in 1979 and strictly enforced in the 1980s. It lasted until 2015, when it was replaced with a two-child policy, and later, in 2021, China allowed families to have three children due to declining birth rates.

# France's Government Incentives for Childbirth have evolved over time, but the major pro-natalist policies began in the 1930s. The most significant modern incentives were reinforced in the early 2000s, including:

# 2003: Expansion of parental leave and financial incentives.
# 2005: The introduction of the "Carte Famille Nombreuse" (Large Family Card), which offers discounts for families with three or more children.
# 2006: Increased cash benefits and tax breaks for families with multiple children to encourage higher birth rates.

# Select data for China and France
df_selected = df[['China', 'France']]

# Plot fertility rates across years
plt.figure(figsize=(9, 5))
plt.plot(df_selected.index, df_selected['China'], label='China', marker='o', linestyle='-')
plt.plot(df_selected.index, df_selected['France'], label='France', marker='s', linestyle='--')

# Formatting
plt.title('Fertility Rate Comparison: China vs. France (1960-2022)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Fertility Rate', fontsize=12)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# INSIGHTS from the Visualization
# China: Sharp fertility rate decline after 1979, stabilizing at lower levels.
# France: Gradual increase after 2003-2006 policies, showing the effectiveness of incentives.
```
